```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
```

```python
stats = pd.read_csv('../recordings/New_recordings/machine/stats.csv')
t = stats['Threshold']
me = stats['Miss_e']
he = stats['Hit_e']
fae = stats['FA_e']

mc = stats['Miss_c']
hc = stats['Hit_c']
fac = stats['FA_c']

stats_e = pd.read_csv('../recordings/New_recordings/machine/stats_subjects_e.csv')
stats_c = pd.read_csv('../recordings/New_recordings/machine/stats_subjects_c.csv') 

mhe = stats_e['Miss']
fahe = stats_e['FA']

mhc = stats_c['Miss']
fahc = stats_c['FA']
```

```python
sns.set_style("white")
sns.set_style("ticks")
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(mc, fac,linewidth = 4,color='red',label='Machine - Chinese')
ax.plot(me, fae,linestyle='dashed',linewidth = 4,color='blue',label='Machine-English')
ax.scatter(x = mhc, y = fahc,alpha = 0.5,s=40,color='red',label='Humans on Chinese')
ax.scatter(x = mhe, y = fahe,alpha = 0.5,s=100,marker= '*',color='blue',label='Humans on English')
ax.scatter(x = np.mean(mhc),y = np.mean(fahc),s=200,marker = 'o',color='red',edgecolors ='black')
ax.scatter(x = np.mean(mhe),y = np.mean(fahe),s=400,marker = '*',color='blue',edgecolors ='black')
plt.grid(True)
ax.minorticks_on()

# Customize the major grid
ax.grid(which='major', linestyle='-', linewidth='0.5', color='gray')
# Customize the minor grid
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='gray')
ax.set_xlim(0,100)
ax.set_ylim(0,100)
sns.despine()
ax.legend(loc='upper right')
plt.xlabel("MISS %")
plt.ylabel("FALSE ALARM %")
plt.tight_layout()
ax.figure.savefig("DET_machine_human.pdf", bbox_inches='tight')
plt.show()
```
